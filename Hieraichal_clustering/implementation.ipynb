```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_scaled=scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
## Apply the PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=iris.target, cmap='viridis')
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20,7))
plt.title("dendogram")
dendrogram = sch.dendrogram(sch.linkage(pca_scaled, method='ward'))
plt.title('dendogram')
plt.xlabel('sample Index')
plt.ylabel('distance')

```

```python
from  sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='complete')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```
