```python
""" independent and dependent features """
x=df[['tip','sex','smoker','day','time','size']]
y=df[['total_bill']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
X_train
```

```python
""" feature encoding"""
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_train['sex']=le1.fit_transform(X_train[['sex']])
X_train['smoker']=le2.fit_transform(X_train[['smoker']])
X_train['time']=le3.fit_transform(X_train[['time']])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])  # Fixed missing parenthesis
```

```python
X_test.head()
```

```python
""" applying one hot encoding"""
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

```

```python
ct=ColumnTransformer(transformers=[('OneHot',OneHotEncoder(drop='first'),[3])],
                                   remainder='passthrough')
```

```python
ct.fit_transform(X_train)  # Correct for multiple columns
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
y_pred=svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)  # Correct way to predict multiple samples
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np

# Ensure y_test and y_pred are 1D numpy arrays
y_test = np.ravel(y_test)  # Flattens to 1D array if needed
y_pred = np.ravel(y_pred)  # Flattens to 1D array if needed

```

```python
print("y_test shape:", y_test.shape)
print("y_pred shape:", y_pred.shape)
y_test = np.ravel(y_test)
y_pred = np.ravel(y_pred)
```

```python
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={ 'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,grid_prediction))
```
